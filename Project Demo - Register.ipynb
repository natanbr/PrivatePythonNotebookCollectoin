```python
#load images, cut the clip
clip1 = VideoFileClip(r"C:\Users\nbraslav\OneDrive\2nd deg project\Demo\Demo_samasung_Coinfig.mp4").subclip(12,23)
clip2 = VideoFileClip(r"C:\Users\nbraslav\OneDrive\2nd deg project\Demo\Demo_LG_Config.mp4").subclip(0,13)

#get the clip size
w,h = clip1.size

#resize the clip and add margin 
clip1 = clip1.resize((w,h)).margin( 2,color=(100,100,100))
clip2 = clip2.resize((w,h)).set_pos('right').margin( 2,color=(100,100,100))

# compose video from the clips
# video = CompositeVideoClip([clip1,clip2], size=(w,h))
```

```python
tfreeze = 11 # Time of the freeze at 0`11
#get the image from the clip
im_freeze = clip1.to_ImageClip(tfreeze)

#show the image
#ipython_display(im_freeze)

#combine to one clip
clip1 =  concatenate_videoclips([clip1,im_freeze.set_duration(2)])
```

```python
#ipython_display(video,fps=24,width=720)
```

```python
txt_s = TextClip("Samsung Nexus", font='Amiri-regular', color='white',fontsize=24)
txt_s = txt_s.set_pos((0,0)).set_duration(13)
txt_col_s = txt_s.on_color(size=(w, txt_s.h + 10), color=(0,0,0), col_opacity=0.9).set_pos((0,0))

txt_l = TextClip("LG G3", font='Amiri-regular', color='white',fontsize=24)
txt_l = txt_l.set_pos((0,0)).set_duration(13)
txt_col_l = txt_l.on_color(size=(w, txt_l.h + 10), color=(0,0,0), col_opacity=0.9).set_pos((w,0))
```

```python
video_txt = CompositeVideoClip([clip1,txt_col_s,clip2,txt_col_l], size=(2*w,h))
```
